```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
# !pip install scikit-surprise
from surprise import Dataset, SVD, NMF, NormalPredictor, KNNBasic
from surprise.model_selection import cross_validate
from surprise.model_selection import GridSearchCV
```

```python
movies_dataset = Dataset.load_builtin(name='ml-100k')
svd = SVD()
nmf = NMF()
normal_predictor = NormalPredictor()
knn = KNNBasic()

results_models = []
for algorithm in [svd, nmf, normal_predictor, knn]:
    results_cv = cross_validate(algorithm, movies_dataset, measures=['RMSE'], cv=5, verbose=False)
    mean_results = {key: np.mean(value) for key, value in results_cv.items()}
    results_models.append(mean_results)


print(results_models)

df = pd.DataFrame({
    'Algorithm': ['SVD', 'NMF', 'Normal Predictor', 'KNN'],
    'Mean_RMSE': [result['test_rmse'] for result in results_models],
    'Fit_Time': [result['fit_time'] for result in results_models],
    'Test_Time': [result['test_time'] for result in results_models]
})
df
```
